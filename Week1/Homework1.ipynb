# แบบฝึกหัดที่ 2 
___

สร้างโปรแกรม Python ที่รับ input เป็นรายการตัวเลขแยกด้วยช่องว่าง แล้วแสดงผลสถิติพื้นฐานเกี่ยวกับชุดข้อมูลดังกล่าว

ปรแกรมนี้เป็นโปรแกรมที่ช่วยในการวิเคราะห์และแสดงผลสถิติพื้นฐานเกี่ยวกับชุดข้อมูลที่ผู้ใช้ป้อนเข้ามา ซึ่งจะคำนวณและแสดงผลสถิติเชิงพรรณาแบบง่ายของชุดข้อมูลดังกล่าว ตัวอย่างเช่น ค่าเฉลี่ย (mean) ค่าสูงสุด (maximum) ค่าต่ำสุด (minimum) และจำนวนสมาชิกทั้งหมดในชุดข้อมูล

โดยผู้ใช้งานจะต้องป้อนชุดข้อมูลโดยเรียงตัวเลขแยกด้วยช่องว่าง แล้วกด Enter เพื่อส่งข้อมูลเข้าสู่โปรแกรม จากนั้นโปรแกรมจะทำการวิเคราะห์และคำนวณสถิติพื้นฐาน และแสดงผลลัพธ์ออกทางหน้าจอ

ตัวอย่างการทำงานของโปรแกรม:

```
โปรแกรมคำนวณสถิติพื้นฐาน
โปรดป้อนชุดข้อมูลที่ต้องการวิเคราะห์ (เลขคั่นด้วยช่องว่าง): 10 15 20 25 30

ผลลัพธ์:
ค่าเฉลี่ย: 20.0
ค่าสูงสุด: 30
ค่าต่ำสุด: 10
จำนวนสมาชิกทั้งหมด: 5
```
