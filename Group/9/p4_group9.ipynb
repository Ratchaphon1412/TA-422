```python
import json
from matplotlib import pyplot as plt
import csv
```

```python
# Load the JSON data from the file
with open('./dataset/P4/in/p4_input_dataset_1.json', 'r') as file:
    data = [json.loads(line) for line in file]
for x in data:
    print(x);
```

```python
# Initialize dictionaries to store engagements for each team
engagements = {
    'teamRed': 0,
    'teamGreen': 0,
    'teamBlue': 0,
    'teamYellow': 0,
    'teamMagenta': 0,
    'teamCyan': 0,
}

# Calculate engagements and update the dictionary
for tweet in data:
    hashtags = tweet.get("Hashtag", "")
    likes = tweet.get("Likes", 0)
    retweets = tweet.get("Retweets", 0)
    if hashtags is not None:
        for hashtag in hashtags.split(", "):
            if hashtag in engagements:
                engagements[hashtag] += likes + retweets
                
# Calculate the total engagements
total_engagements = sum(engagements.values())

# Print the engagements for each team
for team, engagement in engagements.items():
    print(f"{team}: {engagement}")

# Print the total engagements
print(f"Total Engagements: {total_engagements}")
```

```python
# Save the engagements data to a CSV file



with open('p4_output_1.csv', 'w', newline='') as csvfile:
    fieldnames = ['Team', 'Engagements']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    for team, engagement in engagements.items():
        writer.writerow({'Team': team, 'Engagements': engagement})
```

```python
# Filter out teams with no engagements

engagements = {team: engagement for team, engagement in engagements.items() if engagement > 0}

# Check if there are teams with engagements to plot
if engagements:
    # Create a pie chart
    labels = engagements.keys()
    sizes = engagements.values()
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    # Display the pie chart
    plt.title("Engagements by Team")
    fig = plt.gcf()
    fig.set_size_inches(10, 10)
    plt.savefig("./dataset/P4/out/p4_plot_1.png",dpi=100)
    plt.show()
else:
    print("No teams with engagements to plot.")
    
```
