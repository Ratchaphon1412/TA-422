```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from math import ceil
import csv
from math import floor
```

```python
df = pd.read_csv('./dataset/P3/in/p3_input_dataset_1.csv')
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
# hw1 = df.to_numpy(df['HW1'])
# hw2 = df.to_numpy(df['HW2'])
# final = df.to_numpy(df['Final'])
# print(hw1)
# type(hw1)
df['Total'] = df.apply(lambda row: ceil(0.25 * row['HW1'] + 0.25 * row['HW2'] + 0.5 * row['Final']), axis=1)
df.head()
```

```python

mean = df.Total.mean()
standardDeviation = df.Total.std()

cutoff = floor(mean - 2 * standardDeviation)

cutoff
```

```python
df['Grade'] = df['Total'].apply(lambda x: 'S' if x >= cutoff else 'U')
df.to_csv('./dataset/P3/out/p3_output_dataset_1.csv',index=False)
```

```python
total_scores = df['Total']

plt.hist(total_scores, bins=50,range=(0, 100), edgecolor='black', alpha=0.7)

grading_point = cutoff

plt.axvline(grading_point, color='red', linestyle='--', label='Cutoff Line')

plt.title('Distribution of Total Scores with Cutoff Line')
plt.xlabel('Total Score')
plt.ylabel('Frequency')

plt.legend()

fig = plt.gcf()
fig.set_size_inches(10, 10)
plt.savefig("./dataset/P3/out/p3_plot_1.png")
plt.show()
```
