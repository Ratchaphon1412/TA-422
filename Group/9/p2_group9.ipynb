```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
with open('./dataset/P2/in/p2_input_dataset_1.txt','r') as file:
    input1 = file.read()
line = input1.splitlines()

result_line = []
for i in line:
    words = i.split()
    result_lines = ",".join(words)
    result_line.append(result_lines)

result = "\n".join(result_line)

# print(result)
output_csv_file = './dataset/P2/in/p2_input_dataset_1.csv'

with open(output_csv_file, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile, delimiter=',')
    for line in result_line:
        csv_writer.writerow(line.split(','))
        
print(f'Text file converted to CSV: {output_csv_file}')

    
```

```python
df = pd.read_csv('./dataset/P2/in/p2_input_dataset_1.csv')
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
# fill missing values with "mean" of that column using fillna()
mean = round(df.Age.mean())
df.Age = df.Age.fillna(mean) 
df
```

```python
df = df.drop(columns=['PatientID'])
df
```

```python
age_bins = [ 19, 29, 39, 49, 59]  # Customize the bins as needed
age_labels = [ '20-30', '30-40', '40-50', '50-60',]

# Create a new column 'Age Group' with the age groups
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, include_lowest=True)
age_group_distribution = df['Age Group'].value_counts().sort_index()
print(age_group_distribution)
```

```python
output = pd.DataFrame({
    "Age Group": age_labels,
    "Number of Patients": age_group_distribution
})

output = output.reset_index(drop=True)
print(output)
```

```python
output.to_csv('./dataset/P2/out/p2_output_1.csv',index=False)
```

```python
# Create a histogram
plt.bar(output['Age Group'], output['Number of Patients'])
plt.xlabel('Age Group')
plt.ylabel('Number of Patients')
plt.title('Age Distribution of Patients')
plt.xticks(rotation=45)

# Save the histogram
fig = plt.gcf()
fig.set_size_inches(10, 10)
plt.savefig("./dataset/P2/out/p2_plot_1.png")
```
