```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json

# from google.colab import drive
# drive.mount("/content/drive")
```

```python
# ---
# Group No. 11
# นายเจษฎา ทิพลาภรุ่งโรจน์ 6410450397 
# นางสาว เบญญทิพย์ กีรติยาพร 6410450524
# นาย รุสลัน จงผดุงสัตย์ 6410450583
# นางสาว ศรัณย์รัชต์ วงค์น่วม 6410450605
# ---

problem = 4 # ปัญหาที่
data_num = 1 # ชุดข้อมูลที่

# ชุดข้อมูลเก็บใน /in
in_path = "./dataset/P"+str(problem)+"/in/"
data_path = in_path+"p"+str(problem)+"_input_dataset_"+str(data_num)+".json"

# ผลลัพธ์เก็บใน /out
out_path = "./dataset/P"+str(problem)+"/out/"
result_path = out_path+"p"+str(problem)+"_output_"+str(data_num)+".csv"
png_path = out_path+"p"+str(problem)+"_plot_"+str(data_num)+".png"

# อ่านข้อมูล
data = pd.read_json(data_path, lines=True)

# --- เริ่มเขียนโค้ดตรงนี้ ---

# อ่านไฟล์ JSON
with open('./dataset/P4/in/p4_input_dataset_1.json', 'r') as file:
    data = json.load(file)

# นับจำนวน Tweet ที่มี Hashtag และไม่มี Hashtag
hashtag_count = 0
no_hashtag_count = 0

for tweet in data:
    if tweet["Hashtag"] is not None:
        hashtag_count += 1
    else:
        no_hashtag_count += 1

# สร้าง Pie Chart
labels = ['Hashtag', 'No Hashtag']
sizes = [hashtag_count, no_hashtag_count]
colors = ['lightcoral', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # ทำให้กราฟเป็นวงกลม

plt.title('Pie chart of the number of Tweets containing a hashtag in Hashtag \nTotal Engagements: 20')


# เพิ่มรายละเอียดใน Pie Chart
plt.text(0.5, 0.5, f'Hashtag: {hashtag_count}', fontsize=12, ha='center', va='center', color='white')
plt.text(0.5, -0.5, f'No Hashtag: {no_hashtag_count}', fontsize=12, ha='center', va='center', color='white')

plt.show()


# --- สิ้นสุดโค้ดตรงนี้ ---

# เซฟผลลัพธ์
data.to_csv(result_path, index=False)

# แสดงผลลัพธ์
data.head()

# เซฟกราฟ
plt.savefig(png_path, dpi=100)

# แสดงผลกราฟ
plt.show()
```
