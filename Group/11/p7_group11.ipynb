```python
# ---
# Group No. 11
# นายเจษฎา ทิพลาภรุ่งโรจน์ 6410450397 
# นางสาว เบญญทิพย์ กีรติยาพร 6410450524
# นาย รุสลัน จงผดุงสัตย์ 6410450583
# นางสาว ศรัณย์รัชต์ วงค์น่วม 6410450605
# ---

problem = 7 # ปัญหาที่
data_num = 1 # ชุดข้อมูลที่

# ชุดข้อมูลเก็บใน /in
in_path = "./dataset/P"+str(problem)+"/in/"
data_path = in_path+"p"+str(problem)+"_input_dataset_"+str(data_num)+".txt"

# ผลลัพธ์เก็บใน /out
out_path = "./dataset/P"+str(problem)+"/out/"
result_path = out_path+"p"+str(problem)+"_output_"+str(data_num)+".csv"
png_path = out_path+"p"+str(problem)+"_plot_"+str(data_num)+".png"

# อ่านข้อมูล
data = pd.read_csv(data_path)

# --- เริ่มเขียนโค้ดตรงนี้ ---

dates = []
values = []
with open('./dataset/P7/in/p7_input_dataset_1.txt', 'r') as file:
    # Initialize an empty list to store the data
    

    # Read each line from the file
    count = 0
    for line in file:
        # Split the line into date and number using space as the separator
        parts = line.strip()
        if count % 2 == 0 :
            dates.append(parts)
        else :
            values.append(parts)
        count += 1
        

values = [int(val) for val in values]

# Count the frequency of each number
frequency = {}
for val in values:
    if val in frequency:
        frequency[val] += 1
    else:
        frequency[val] = 1

unique_values = list(frequency.keys())
counts = list(frequency.values())

# Create a horizontal bar chart to show the numbers and their frequencies
plt.figure(figsize=(8, 8))
plt.barh(unique_values, counts)
plt.ylabel('Value')
plt.xlabel('Frequency')
plt.title('Number Frequency (Horizontal Bar Chart)')
plt.xticks(np.arange(0, max(counts) + 1, 1))
plt.grid(axis='x')

plt.show()


# --- สิ้นสุดโค้ดตรงนี้ ---

# เซฟผลลัพธ์
data.to_csv(result_path, index=False)

# แสดงผลลัพธ์
data.head()

# เซฟกราฟ
plt.savefig(png_path, dpi=100)

# แสดงผลกราฟ
plt.show()
```
