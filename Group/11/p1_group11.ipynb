```python
# ---
# Group No. 11
# นายเจษฎา ทิพลาภรุ่งโรจน์ 6410450397 
# นางสาว เบญญทิพย์ กีรติยาพร 6410450524
# นาย รุสลัน จงผดุงสัตย์ 6410450583
# นางสาว ศรัณย์รัชต์ วงค์น่วม 6410450605
# ---

problem = 1 # ปัญหาที่
data_num = 1 # ชุดข้อมูลที่

# ชุดข้อมูลเก็บใน /in
in_path = "./dataset/P"+str(problem)+"/in/"
data_path = in_path+"p"+str(problem)+"_input_dataset_"+str(data_num)+".csv"

# ผลลัพธ์เก็บใน /out
out_path = "./dataset/P"+str(problem)+"/out/"
result_path = out_path+"p"+str(problem)+"_output_"+str(data_num)+".csv"
png_path = out_path+"p"+str(problem)+"_plot_"+str(data_num)+".png"

# อ่านข้อมูล
data = pd.read_csv(data_path)

# --- เริ่มเขียนโค้ดตรงนี้ ---
# อ่านไฟล์ CSV
with open('./dataset/P1/in/p1_input_dataset_1.csv', 'r') as f:
  reader = csv.reader(f)

  # กำหนดชื่อคอลัมน์
  headers = next(reader)

  # เก็บข้อมูลในรายการ
  dates = []
  open = []
  high = []
  low = []
  closing_prices = []
  moving_averages = []
  volume = []

  for row in reader:
    open.append(float(row[headers.index('Open')],))
    dates.append(row[headers.index('Date')])
    closing_prices.append(float(row[headers.index('Close')],))
    volume.append(float(row[headers.index('Volume')],))

plt.figure(figsize=(20, 6))
# สร้างกราฟ
plt.plot(dates, open, label='Opening Price')
plt.plot(dates, closing_prices, label='Closing Price')
#plt.fill_between(dates, 0, volume, label='Volume', alpha=0.5)
# ตั้งค่าตัวเลือกกราฟ
plt.title('Stock Price Data')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=60)  # หมุนแกน x เพื่อแสดงวันที่ให้เหมาะสม
plt.tight_layout()  # ปรับระยะห่างในกราฟ
plt.show()



# --- สิ้นสุดโค้ดตรงนี้ ---

# เซฟผลลัพธ์
data.to_csv(result_path, index=False)

# แสดงผลลัพธ์
data.head()

# เซฟกราฟ
plt.savefig(png_path, dpi=100)

# แสดงผลกราฟ
plt.show()
```
