```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from faker import Faker
import json

# from google.colab import drive
# drive.mount("/content/drive")
```

```python
# ---
# Group No. 3
# 1.  ,  6410450061
# 2.,, 6410451881
# 3. , 6410450567

# ---
group = 3
problem = 4 # 
data_num = 1 # 

#  /in
in_path = "./dataset/P"+str(problem)+"/in/"
data_path = in_path+"p"+str(problem)+"_input_dataset_"+str(data_num)+".json"

#  /out
out_path = "./dataset/P"+str(problem)+"/out/"
result_path = out_path+"p"+str(problem)+"_group"+str(group)+"-output"+str(data_num)+".csv"
png_path = out_path+"p"+str(problem)+"_group"+str(group)+"-plot"+str(data_num)+".png"

# 
data = pd.read_json(data_path, lines=True)

# ---  ---
df = pd.DataFrame(data)

#  (NaN)
df = df.dropna()

#  Hashtag  'str.split(', ')'
df['Hashtag'] = df['Hashtag'].str.split(', ')

#  Hashtag  engagement  Hashtag
selected_hashtags = ['teamRed', 'teamGreen', 'teamCyan', 'teamBlue', 'teamMagenta']
hashtags_data = []

for hashtag in selected_hashtags:
    hashtag_data = {
        'Hashtag': hashtag,
        'Likes': 0,
        'Retweets': 0
    }

    #  engagement  Hashtag
    for index, row in df.iterrows():
        if hashtag in row['Hashtag']:
            hashtag_data['Likes'] += row['Likes']
            hashtag_data['Retweets'] += row['Retweets']

    hashtag_data['TotalEngagement'] = hashtag_data['Likes'] + hashtag_data['Retweets']
    hashtags_data.append(hashtag_data)

#  DataFrame  Hashtag
result_df = pd.DataFrame(hashtags_data)

#  Hashtag  TotalEngagement 
result_df = result_df.sort_values(by='TotalEngagement', ascending=False)

# 
print(result_df)

import matplotlib.pyplot as plt

#
team_engagements = result_df.groupby('Hashtag')[['TotalEngagement']].sum()

#
team_engagements = team_engagements[team_engagements['TotalEngagement'] > 0]

#
plt.figure(figsize=(8, 8))
plt.pie(team_engagements['TotalEngagement'], labels=team_engagements.index, autopct='%1.1f%%', startangle=140)
plt.title('Engagements Distribution by Campaign-Related Hashtags Total Engagements :212')
plt.axis('equal')  # 

# ---  ---

# 
result_df.to_csv(result_path, index=False)

# 
#data.head()

# 
plt.savefig(png_path, dpi=100)

# 
plt.show()
```
