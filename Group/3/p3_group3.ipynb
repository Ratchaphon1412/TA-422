```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# from google.colab import drive
# drive.mount("/content/drive")
```

```python
# ---
# Group No. 3
# 1. สุภัคนภา ,ใจเย็น  6410450061
# 2.,สโรชา,อัศวพิชญโชติ 6410451881
# 3. มงคล,คงเสมา 6410450567

# ---
group = 3
problem = 3 # ปัญหาที่
data_num = 1# ชุดข้อมูลที่

# ชุดข้อมูลเก็บใน /in
in_path = "./dataset/P"+str(problem)+"/in/"
data_path = in_path+"p"+str(problem)+"_input_dataset_"+str(data_num)+".csv"

# ผลลัพธ์เก็บใน /out
out_path = "./dataset/P"+str(problem)+"/out/"
result_path = out_path+"p"+str(problem)+"_group"+str(group)+"-output"+str(data_num)+".csv"
png_path = out_path+"p"+str(problem)+"_group"+str(group)+"-plot"+str(data_num)+".png"

# อ่านข้อมูล
data = pd.read_csv(data_path)

# --- เริ่มเขียนโค้ดตรงนี้ ---
import math
df = data
df["Total"] = (0.25 * df["HW1"] + 0.25 * df["HW2"] + 0.5 * df["Final"]).apply(math.ceil)
mean = df["Total"].mean()
std_deviation = df["Total"].std()
cutoff = math.floor(mean - 2 * std_deviation)

#ตรวจสอบและกำหนดเกรด S หรือ U สำหรับแต่ละคะแนนรวม
grades = []
for total_score in df['Total']:
    if total_score >= cutoff:
        grades.append('S')
    else:
        grades.append('U')

#เพิ่มคอลัมน์เกรดลงใน DataFrame
df['Grade'] = grades

sorted_df = df.sort_values(by='Total', ascending=False)

#แสดงผลลัพธ์
print(sorted_df)
print(cutoff)


import matplotlib.pyplot as plt

#สร้างแผนภูมิแท่ง (histogram) แสดงการกระจายตัวของคะแนนรวม
plt.figure(figsize=(10, 10))  # กำหนดขนาดของแผนภูมิ

#สร้าง histogram
#พลอต histogram ของคะแนนรวม 'Total'
plt.hist(sorted_df["Total"], bins=range(0, 101,1), color='lightgray', edgecolor='black')


#วาดเส้น cutoff (แนวตั้ง สีแดง)
plt.axvline(x=cutoff, color='red', linewidth=2, linestyle='solid')
#แทรกข้อความบนเส้นcutoff
plt.text(cutoff+0.5,2.5, str(cutoff), color='red', fontsize=12, verticalalignment='bottom')

#กำหนดชื่อแกน x, y และชื่อแผนภูมิ
plt.xlabel('Total score')
plt.ylabel('Number of students')
plt.title('Total Score Distribution')
plt.legend()  # แสดงคำอธิบายแผนภูมิ

#กำหนดแกน Y และ แกนX
plt.yticks([0, 5,])
plt.xticks(range(0, 101,10))

#เซฟผลลัพธ์
sorted_df.to_csv(result_path, index=False)

#แสดงผลลัพธ์
sorted_df.head()

#เซฟกราฟ # บันทึกแผนภูมิในไฟล์ PNG ขนาด 1000x1000 พิกเซล
plt.savefig(png_path, dpi=100) # dpi = 1000 / 10 = 100

#แสดงผลกราฟ
plt.show()
```
