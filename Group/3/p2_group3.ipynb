```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# from google.colab import drive
# drive.mount("/content/drive")
```

```python
# ---
# Group No. 3

# 1. สุภัคนภา ใจเย็น  6410450061
# 2.,สโรชาอัศวพิชญโชติ 6410451881
# 3. มงคล คงเสมา 6410450567
# ---
group = 3
problem = 2 # ปัญหาที่
data_num = 1 # ชุดข้อมูลที่

# ชุดข้อมูลเก็บใน /in
in_path = "./dataset/P"+str(problem)+"/in/"
data_path = in_path+"p"+str(problem)+"_input_dataset_"+str(data_num)+".txt"

# ผลลัพธ์เก็บใน /out
out_path = "./dataset/P"+str(problem)+"/out/"
result_path = out_path+"p"+str(problem)+"_group"+str(group)+"-output"+str(data_num)+".csv"
png_path = out_path+"p"+str(problem)+"_group"+str(group)+"-plot"+str(data_num)+".png"

# อ่านข้อมูล
data = pd.read_csv(data_path, delimiter='\s+')


# --- เริ่มเขียนโค้ดตรงนี้ ---
# สร้าง DataFrame จากข้อมูล
df = pd.DataFrame(data)

# จัดการค่าที่ขาดหายไป
df['Age'].fillna(np.nan, inplace=True)

# ตรวจสอบประเภทข้อมูล
# ให้ df เป็น DataFrame ที่มีคอลัมน์ที่ต้องการแปลงเป็น integer คือ 'Age'
# แทนที่ค่า NA ด้วยค่าเฉลี่ยของคอลัมน์
df['Age'] = df['Age'].fillna(df['Age'].mean()).astype(int)

# กำหนด bins หรือช่วงอายุ
age_bins = [20, 30, 40, 50, 60, 70, 80, 90]

# ใช้ pd.cut() เพื่อแบ่งกลุ่มอายุ
age_groups = pd.cut(df['Age'], bins=age_bins)

# นับจำนวนของแต่ละกลุ่มอายุ
age_counts = age_groups.value_counts().reset_index()
age_counts.columns = ['Age Group', 'Number of Patients']

# กรองแถวที่ Number of Patients ไม่เท่ากับ 0
age_counts_filtered = age_counts[age_counts['Number of Patients'] != 0]

print(age_counts_filtered)

# จัดเรียงข้อมูลตาม Age Group จากน้อยไปมาก
age_counts_filtered_m = age_counts_filtered.sort_values(by='Age Group')

# สร้างกราฟแผนภูมิแท่ง (histogram)
plt.figure(figsize=(8, 6))
plt.bar(age_counts_filtered_m['Age Group'].astype(str), age_counts_filtered_m['Number of Patients'], color='skyblue')

# กำหนดชื่อแกน x และ y และชื่อกราฟ
plt.xlabel('Age Group')
plt.ylabel('Number of Patients')
plt.title('Number of Patients by Age Group')

# แสดงกราฟ
plt.xticks(rotation=0)  # หมุนชื่อกลุ่มอายุเพื่อให้สามารถอ่านได้ง่ายขึ้น
plt.tight_layout()

# --- สิ้นสุดโค้ดตรงนี้ ---

# เซฟผลลัพธ์
age_counts_filtered.to_csv(result_path, index=False)

# แสดงผลลัพธ์
#data.head()

# เซฟกราฟ
plt.savefig(png_path, dpi=100)

# แสดงผลกราฟ
plt.show()

#
```
