```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./dataset/P1/in/p1_input_dataset_1.csv')
df.head()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['30MA'] = df['Close'].rolling(window=30).mean()
df['30MA']
```

```python
df['Volatility'] = df['Close'].rolling(window=30).std()
df['Volatility']
```

```python
plt.figure(figsize=(10, 10))
plt.plot(df['Date'], df['Close'])
plt.plot(df['Date'], df['30MA'])
plt.fill_between(df['Date'], df['30MA'] - df['Volatility'], df['30MA'] + df['Volatility'], alpha=0.2)
fig = plt.gcf()
fig.set_size_inches(10, 10)
plt.savefig("./dataset/P1/out/p1_plot_1.png")
```

```python

df.to_csv('./dataset/P1/out/p1_output_1.csv', index=False)
```
