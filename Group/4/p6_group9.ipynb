```python
import pandas as pd

df = pd.read_csv('./dataset/P6/in/p6_input_dataset_1.csv')

df['royalty_number'] = pd.to_numeric(df['royalty_number'], errors='coerce')

df['date'] = pd.to_datetime(df['date'])

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

df['membership'] = df['royalty_number'].notna()

result = df.groupby(['year', 'month', 'membership'])['total_sale'].sum().reset_index()

result.to_csv('./dataset/P6/out/p6_output_dataset_1.csv', index=False)
```

```python
import calendar
import matplotlib.pyplot as plt

result['month'] = result['month'].apply(lambda x: calendar.month_name[int(x)])
pivot_df = result.pivot(index=['year', 'month'], columns='membership', values='total_sale')

plt.figure(figsize=(10, 10))
pivot_df.plot(kind='bar')
plt.title('Total Sales by Royalty for Each Munth')
plt.xlabel('Year, Month')
plt.ylabel('Total Sales')
plt.legend(['Non Royalty', 'Royalty'], title='Royalty Status')
plt.xticks(rotation=45)
plt.tight_layout()

fig = plt.gcf()
fig.set_size_inches(10, 10)
plt.savefig('./dataset/P6/out/p6_plot_1.png',dpi=100)

plt.show()
```
